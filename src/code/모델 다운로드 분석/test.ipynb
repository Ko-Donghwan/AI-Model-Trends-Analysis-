```python
import pandas as pd
```

```python
df = pd.read_csv('./csv/Model_2025-02-11-crawling.csv')
```

```python
df.iloc[0,:]
```

```python
# 1Day~30Day 누적합 계산
df.iloc[0, 2:] = df.iloc[0, 2:].astype(int).cumsum()

print(df)
```

```python
df.iloc[:, 2:] = df.iloc[:, 2:].astype(int).cumsum(axis=1)
```

```python
df.to_csv("./csv/Model_2025-02-11-cumsum.csv", index=False)
```
